```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
ds = pd.read_csv('/Users/dr_sync/Downloads/Mini Project/dataset/diabetes.csv') 
```

```python
x = ds.drop(columns = 'Outcome', axis=1)
y = ds['Outcome']
```

```python
xtr, xte, ytr, yte = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=1)
```

```python
model = svm.SVC(kernel='linear')
model.fit(xtr, ytr)
```

```python
xtrp = model.predict(xtr)
trda = accuracy_score(xtrp, ytr)
print('Accuracy score of the training data : ', trda)
```

```python
xtep = model.predict(xte)
teda = accuracy_score(xtep, yte)
print('Accuracy score of the test data : ', teda)
```

```python
id = (5,166,72,19,175,25.8,0.587,51)
numArr = np.asarray(id)
reshapedArr = numArr.reshape(1,-1)

p = model.predict(reshapedArr)
if (p[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
