```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
ds = pd.read_csv('/Users/dr_sync/Downloads/Mini Project/dataset/parkinsons.csv')
```

```python
x = ds.drop(columns=['name','status'], axis=1)
y = ds['status']
```

```python
xtr, xte, ytr, yte = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
model = svm.SVC(kernel='linear')
model.fit(xtr, ytr)
```

```python
xtrp = model.predict(xtr)
trda = accuracy_score(ytr, xtrp)
print('Accuracy score of training data : ', trda)
```

```python
xtep = model.predict(xte)
teda = accuracy_score(yte, xtep)
print('Accuracy score of test data : ', teda)
```
