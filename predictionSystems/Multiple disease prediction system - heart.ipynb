```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ds = pd.read_csv('/Users/dr_sync/Downloads/Mini Project/dataset/heart.csv')
```

```python
x = ds.drop(columns='target', axis=1)
y = ds['target']
```

```python
xtr, xte, ytr, yte = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(xtr, ytr)
```

```python
xtrp = model.predict(xtr)
trda = accuracy_score(xtrp, ytr)
print('Accuracy on Training data : ', trda)
```

```python
xtep = model.predict(xte)
teda = accuracy_score(xtep, yte)
print('Accuracy on Test data : ', teda)
```
